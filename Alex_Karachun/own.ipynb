```python
df = pd.read_csv("../../slovo_full/annotations.csv", sep='\t')
```

```python
df = df[df['height'] > df['width']]
```

```python
df.shape
```

```python
aspect_ratios = df.groupby(["height", "width"]).size().reset_index(name="count")

# Сортировка по количеству (опционально)
aspect_ratios = aspect_ratios.sort_values(by="count", ascending=False)

print(aspect_ratios)
```

```python
aspect_ratios['/'] = aspect_ratios['height'] / aspect_ratios['width']
```
